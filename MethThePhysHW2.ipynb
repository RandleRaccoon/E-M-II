```python
import numpy as np
x=np.linspace(0,1,100)
#print(x)
for i in range(0,100):
  #x=np.linspace(0,1,100)
  E1=1+x[i]*1/2+x[i]**2*3/8+x[i]**3*5/16
  E2=1/np.sqrt(1-x[i])
  P=E1/E2

  print(x[i],E1,E2,P)
```
