```python
import numpy as np
import matplotlib.pyplot as plt
# Defines the cartesian grid
x,y = np.meshgrid(np.linspace(-5,5,11),np.linspace(-5,5,11))
#Define the field in the cartesian grid
a,b=(5,7)
phi = np.arctan2(y, x)
r = np.sqrt(x**2 + y**2)
#a = 2*b
u=b*np.cos(phi)-a*np.sin(phi)
v=b*np.sin(phi)+a*np.cos(phi)
#Create figure to plot things in:
fig = plt.figure(figsize = (7, 7))
#Plot the vector field
plt.quiver(x,y,u,v,color='red')
#Plot the streamlines
plt.streamplot(x, y, u, v, density = 1.0,color='blue')
plt.xlabel('x axis',fontsize=20)
plt.ylabel('y axis',fontsize=20)
plt.grid(True)
plt.show()


u=b*r**2*np.cos(phi)-a*r**2*np.sin(phi)
v=b*r**2*np.sin(phi)+a*r**2*np.cos(phi)
#Create figure to plot things in:
fig = plt.figure(figsize = (7, 7))
#Plot the vector field
plt.quiver(x,y,u,v,color='red')
#Plot the streamlines
plt.streamplot(x, y, u, v, density = 1.0,color='blue')
plt.xlabel('x axis',fontsize=20)
plt.ylabel('y axis',fontsize=20)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Defines the cartesian grid
x,y = np.meshgrid(np.linspace(-5,5,11),np.linspace(-5,5,11))
#Define the field in the cartesian grid
a,b=(5,7)
phi = np.arctan2(y, x)
r = np.sqrt(x**2 + y**2)
#a = 2*b
u=b*(1+a**2/r**2)-2*b*a**2*x*r**(-3)*np.cos(phi)
v=-np.sin(phi)*2*b*a**2*x*r**(-3)
#Create figure to plot things in:
fig = plt.figure(figsize = (7, 7))
#Plot the vector field
plt.quiver(x,y,u,v,color='red')
#Plot the streamlines
plt.streamplot(x, y, u, v, density = 1.0,color='blue')
plt.xlabel('x axis',fontsize=20)
plt.ylabel('y axis',fontsize=20)
plt.grid(True)
plt.show()
```
